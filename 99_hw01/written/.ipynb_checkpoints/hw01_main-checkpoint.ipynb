```python
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
```

Problem 3.1.

```python
def generate_data(num_samples, margin, w_star, seed=12345):
    x_train = []
    y_train = []
    cnt = 0
    
    np.random.seed(seed)
    # uniform_dist = stats.uniform(loc=[-1,-1], scale=[1,1])
    while cnt < num_samples:
        # x_i = uniform_dist.rvs(size=(1,2))
        x_i = np.random.uniform(low=[-1,-1], high=[1,1], size=(1,2))
        # print(x_i.shape)
        x_i_norm = np.matmul(x_i, x_i.T)
        # print(x_i_norm)
        logit = np.matmul(x_i, w_star)
        if np.abs(logit) >= margin and x_i_norm <= 1:
            y_i = np.sign(logit)
            # print(logit)
            # print(x_i, y_i)
            x_train.append(x_i)
            y_train.append(y_i)
            cnt += 1
        # else:
        #     print(logit)
        #     stop
            
    x_train = np.vstack(x_train)
    y_train = np.vstack(y_train)
    # print(x_train[:10])
    # print(x_train.shape)
    # print(y_train.shape)
    return x_train, y_train

# parameters
num_samples = 1000
margin_l = 1/32
w_star = np.array([1/np.sqrt(2), 1/np.sqrt(2)])
print(f"w* = {w_star}")

# generate data
x_train, y_train = generate_data(num_samples, margin_l, w_star, seed=12345) 

# plot
# print(np.sum(y_train>=0)/num_samples)
ypos_idxes = np.where(y_train >= 0)
yneg_idxes = np.where(y_train < 0)

fig, axes = plt.subplots(1, 1, figsize=(3,3))
axes = [axes]
axes[0].scatter(x_train[ypos_idxes,0], x_train[ypos_idxes,1], color="blue", label=1, s=10, marker="o")
axes[0].scatter(x_train[yneg_idxes,0], x_train[yneg_idxes,1], color="green", label=-1, s=10, marker="x")
axes[0].legend(loc="upper right")
plt.show()
```

Problem 3.2.

```python
def sim_perceptron(x_train, y_train, w_star, num_samples):
    w_trained = np.zeros(2)
    # print(w_trained)
    cnt_mistakes = 0
    w_list1 = []
    w_list2 =  []
    for i in range(num_samples):
        # print(x_train[i,:].shape)
        # print(w_trained.shape)
        # stop
        logit = np.dot(x_train[i,:], w_trained)
        y_hat_i = np.sign(logit)
        if y_hat_i != y_train[i]:
            w_trained = w_trained + y_train[i]*x_train[i,:]
            cnt_mistakes += 1
            w_list1.append(np.dot(w_trained.T, w_star))
            w_list2.append(np.dot(w_trained, w_trained))
        
    return cnt_mistakes, w_trained, w_list1, w_list2
    
cnt_mistakes, w_trained, w_list1, w_list2 = sim_perceptron(x_train, y_train, w_star, num_samples)
print(f"Num mistakes, M = {cnt_mistakes}")
print(f"Final weight, w = {w_trained}")
# print(w_list1[:10])
# print(w_list2[:10])


fig, axes = plt.subplots(1, 1, figsize=(5,5))
axes = [axes]

m_range = np.arange(1, cnt_mistakes+1)
axes[0].plot(m_range, w_list1, color="blue", linestyle='-', label=f"$\langle w^m, w^* \\rangle $")
axes[0].plot(m_range, m_range*margin_l, color="blue", linestyle='--', label="$\gamma*m$")
axes[0].plot(m_range, w_list2, color="green", linestyle='-', label="$\langle w^m, w^m \\rangle $")
axes[0].plot(m_range, np.sqrt(m_range), color="green", linestyle='--', label="$\sqrt{m}$")

axes[0].legend(loc="upper right")
plt.show()
```

From the plot we can verify that 
$$
\forall m \in \{0,1,\ldots,M\}, \langle w^m, w^* \rangle \ge \gamma m ~\text{and}~ \|w^m\| \le \sqrt{m}
$$

Problem 3.3.

```python
# parameters
num_datasets = 10
num_samples = 100

margin_list = [2**(-i) for i in range(1,7)]
m_avg_list = []
for l, margin_l in enumerate(margin_list):
    m_l = 0 
    # print(f"\n->l={l}, margin={margin_l}")
    for d in range(num_datasets):
        x_train_ld, y_train_ld = generate_data(num_samples, margin_l, w_star, seed=10000+l+d)
        m_ld, _, _, _ = sim_perceptron(x_train_ld, y_train_ld, w_star, num_samples)
        # print(m_ld)
        m_l += m_ld
    # print(m_l/num_datasets)
    m_avg_list.append(m_l/num_datasets)

fig, axes = plt.subplots(1, 1, figsize=(5,5))
axes = [axes]

m_range = np.arange(1, cnt_mistakes+1)
axes[0].plot(margin_list, m_avg_list, color="blue", linestyle='-', label="$\\bar{M}_{\gamma}$")
axes[0].plot(margin_list, 1/(np.array(margin_list)**2), color="green", linestyle='--', label=f"$g(\gamma)=1/ \gamma^2$")

axes[0].legend(loc="upper right")
plt.show()
```

The plot of $\bar{M}_{\gamma}$ always below the plot of $g(\gamma) = \frac{1}{\gamma^2}$. This result confirms the Perceptron convergence theorem that the Perceptron algorithm always makes at most $\frac{1}{\gamma^2}$ mistakes throughout the process.
