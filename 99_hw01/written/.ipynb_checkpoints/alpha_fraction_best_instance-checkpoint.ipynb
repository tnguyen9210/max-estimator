```python
import numpy as np
import scipy.stats as stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

np.set_printoptions(precision=4, suppress=True)

import matplotlib.pyplot as plt
```

```python
# parameters 
mu_opt = 4.0
mu_subopt = 1.0
sigma = 1

num_opt_arms = 10
num_sub_arms = 0
```

```python
def run_trial(trial_N, trial_K_opt, trial_K_subopt, num_trials):
    print(f"\n-> trial_N: {trial_N}, trial_K_opt: {trial_K_opt}, trial_K_subopt: {trial_K_subopt},")
    
    # parameters
    # set \gamma = \sqrt{\frac{4}{N}\log\(\frac{K^2}{\delta})}
    trial_K = trial_K_opt + trial_K_subopt
    gamma = np.sqrt(4/trial_N*np.log(trial_K**2))
    
    #
    mse_tme_est = 0
    mse_max_est = 0
    mse_avg_est = 0
    for tidx in range(num_trials): 
        # for each arm, generate a sample of size N
        muhat_all = []
        for aidx in range(trial_K_opt):
            sample_a = np.random.normal(mu_opt, sigma, size=(trial_N))
            muhat_a = np.sum(sample_a)/trial_N
            muhat_all.append(muhat_a)
        
        for aidx in range(trial_K_subopt):
            sample_a = np.random.normal(mu_subopt, sigma, size=(trial_N))
            muhat_a = np.sum(sample_a)/trial_N
            muhat_all.append(muhat_a)
            
        print(len(muhat_all))
        print(trial_K)
        assert(len(muhat_all) == trial_K)

        # compute muhat_max
        muhat_max = np.max(muhat_all)
        # print(f"muhat_all: {muhat_all}")
        # print(f"muhat_max: {muhat_max:0.4f}")

        # filter arms with the elimination condition
        muhat_survive = []
        cnt = 0
        for muhat_a in muhat_all:
            if muhat_max - muhat_a <= 2*gamma:
                muhat_survive.append(muhat_a)
                cnt += 1
        
        # compute muhat_est
        muhat_tme = np.average(muhat_survive)
        # print(f"muhat_survive: {np.array(muhat_survive)}")
        # print(f"muhat_tme: {muhat_tme:0.4f}")
        
        # compute mse_est
        mse_tme_est += (muhat_tme - mu_opt)**2
        mse_max_est += (muhat_max - mu_opt)**2
        mse_avg_est += (np.average(muhat_all) - mu_opt)**2
    
    return [mse_tme_est/num_trials, mse_max_est/num_trials, mse_avg_est/num_trials]

res = run_trial(50, 1, 4, 10)
print(np.array(res))
res = run_trial(50, 1, 4, 10)
print(np.array(res))
res = run_trial(1000, 1, 4, 10)
print(np.array(res))
```

```python
np.random.seed(12345)

fixed_K = 5
num_trials = 100

# num_samples_list = [10**i for i in range(5)]
num_samples_list = [5, 10, 50, 100, 250, 500, 1000, 1250, 1500]

mse_tme_est_array = np.zeros(len(num_samples_list))
mse_max_est_array = np.zeros(len(num_samples_list))
mse_avg_est_array = np.zeros(len(num_samples_list))
mse_bound_array = np.zeros(len(num_samples_list))
for nidx, N in enumerate(num_samples_list):
    # print(f"\n-> N = {N}")
    mse_ests = run_trial(N, fixed_K, num_trials)
    mse_tme_est_array[nidx] = mse_ests[0]
    mse_max_est_array[nidx] = mse_ests[1]
    mse_avg_est_array[nidx] = mse_ests[2]
    mse_bound_array[nidx] = 4*4/(fixed_K*N)
    
# plot
fig, axes = plt.subplots(1, 1, figsize=(3,3))
axes = [axes]
axes[0].plot(num_samples_list, mse_bound_array, color="red", label=f"mse_bound")
axes[0].plot(num_samples_list, mse_max_est_array, color="green", label=f"mse_max_est")
axes[0].plot(num_samples_list, mse_avg_est_array, color="brown", label=f"mse_avg_est")
axes[0].plot(num_samples_list, mse_tme_est_array, color="blue", label=f"mse_tme_est")
axes[0].legend(loc="upper right")
axes[0].set_xscale('log')
plt.show()
```

### Analyze decay rate in term of K

```python
np.random.seed(12345)

fixed_N = 1000
num_trials = 100

# num_arms_list = [10**i for i in range(5)]
num_arms_list = [5, 10, 50, 100, 250, 500, 1000, 1250, 1500]

mse_tme_est_array = np.zeros(len(num_arms_list))
mse_max_est_array = np.zeros(len(num_arms_list))
mse_avg_est_array = np.zeros(len(num_arms_list))
mse_bound_array = np.zeros(len(num_arms_list))
for nidx, K in enumerate(num_arms_list):
    # print(f"\n-> N = {N}")
    mse_ests = run_trial(fixed_N, K, num_trials)
    mse_tme_est_array[nidx] = mse_ests[0]
    mse_max_est_array[nidx] = mse_ests[1]
    mse_avg_est_array[nidx] = mse_ests[2]
    mse_bound_array[nidx] = 4*4/(K*fixed_N)
    
# plot
fig, axes = plt.subplots(1, 1, figsize=(3,3))
axes = [axes]
axes[0].plot(num_arms_list, mse_bound_array, color="red", label=f"mse_bound")
axes[0].plot(num_arms_list, mse_max_est_array, color="green", label=f"mse_max_est")
axes[0].plot(num_arms_list, mse_avg_est_array, color="brown", label=f"mse_avg_est")
axes[0].plot(num_arms_list, mse_tme_est_array, color="blue", label=f"mse_tme_est")
axes[0].legend(loc="upper right")
axes[0].set_xscale('log')
plt.show()
```
