```python
num_steps_train = 10000

max_eps = 1.0
min_eps = 0.05
decay_rate = 0.0001
# min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode)

eps_ary_poly =  np.zeros(num_steps_train)
eps_ary_exp =  np.zeros(num_steps_train)
for i in range(0, num_steps_train):
    eps_ary_poly[i] = 1.0/np.sqrt(i+1.0)
    eps_ary_exp[i] = min_eps + (max_eps - min_eps)*np.exp(-decay_rate*(i+1.0))

fig, axes = fig, axes = plt.subplots(
        nrows=1, ncols=1, sharex=True, sharey=False, figsize=(8,8))
axes = [axes]

x_ary = np.arange(0, num_steps_train)
axes[0].plot(x_ary, eps_ary_poly, label="eps_poly")
axes[0].plot(x_ary, eps_ary_exp, label="eps_poly")

plt.show()
```
