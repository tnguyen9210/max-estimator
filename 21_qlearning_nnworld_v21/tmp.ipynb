```python
import numpy as np
```

```python
num_data = 10000
a_ary = np.random.choice([10, -12], size=num_data)
a_mean = np.mean(a_ary)
a_std = np.std(a_ary, ddof=1)
print(a_mean)
print(a_std)
```
